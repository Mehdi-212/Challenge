```python
#importation des librairie
import pandas as pd
import numpy as np
```

```python
#lecture du fichier et selection des colonnes qui nous interresses
df = pd.read_table("marathon.txt", names=["ville","annee","temps","secondes","estPARIS","estBERLIN"])
df
```

```python
#permet de changer le Nan en 1 pour paris et berlin 

for i in range (len(df.ville)):
    if df.ville[i] == "PARIS":
        df.estPARIS[i] = 1
        df.estBERLIN[i] = 0
        #print(sum(df.estPARIS))

    elif df.ville[i] == "BERLIN":
        df.estPARIS[i] = 0
        df.estBERLIN[i] = 1
        #print(sum(df.estBERLIN))

    else:
        df.estPARIS[i] = 0
        df.estBERLIN[i] = 0
        
print(df.head (n=3))
```

```python
#x_transpose = 3 lignes
x_transpose= np.matrix([df.annee, df.estPARIS,df.estBERLIN])
x_transpose
```

```python
#T PERMET DE FAIRE LA TRANSPOSE (x = 3 colonnes)
x = x_transpose.T
x
```

```python
# multiplication entre x et sa transposer
a = x_transpose*x
a
```

```python
# inverse
a1 = np.linalg.inv(a)#commande numpy qui permet de fait l'inverse
a1
```

```python
# inverse * transposer
a2 = a1*x_transpose
a2
```

```python
#initialisation de y
y =np.matrix([df.secondes]).T
y
```

```python
#inverse * y
a3 =a2 * y
a3
#ou cette methode
#((x.T)*x).I*x.T*
print ('on trouve que a0 vaut',a3[0],"que a1 vaut",a3[1],"que a2 vaut",a3[2])
```
